```python
# import dependencies

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the tips data from seaborn dataset

data = sns.load_dataset('tips')
```

```python
# Display first 5 row of the data

data.head()
```

```python
# describe the data like total number of row mean,standrad deviation, min,max, median etc.

data.describe()
```

```python
# scatter plot 
sns.relplot(x="total_bill", y="tip", data=data);
```

```python
sns.relplot(x = 'total_bill',y='tip',hue = 'smoker',data = data)
```

```python
# catagorical plot
sns.catplot(x = 'day',y = 'total_bill',data = data)
```

```python
# jitter 
sns.catplot(x = 'day',y = 'total_bill',jitter = False,data = data)
```

```python
sns.catplot(x = 'smoker',y='tip',data = data)
```

```python
sns.catplot(x = 'smoker',y='tip',order = ['No','Yes'],data = data)
```
